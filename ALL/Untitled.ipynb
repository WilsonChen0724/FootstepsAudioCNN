```python
import os
import tensorflow as tf
from matplotlib import pyplot as plt
import tensorflow_io as tfio
```

```python
瑋階梯_file = os.fsencode(os.path.join('noise_襪子 1.wav')) #注意這個為相對路徑如果把程式碼進行移動這裡也要跟著改
頻階梯_file = os.fsencode(os.path.join('襪子 1.wav'))
```

```python
def load_wav_16000_mono(filename):
    # Load encoded wav file
    file_contents = tf.io.read_file(filename)
    # Decode wav (tensors by channels) 
    wav, sample_rate = tf.audio.decode_wav(file_contents, desired_channels=1)
    wav = tf.squeeze(wav,axis = -1)                                # Removes trailing axi5343wav, axis=-1)
    sample_rate = tf.cast(sample_rate, dtype=tf.int64)
    # Goes from 44100Hz to 16000hz - amplitude of the audio signal
    wav = tfio.audio.resample(wav, rate_in=sample_rate, rate_out=16000) #將採樣頻率轉成16000赫茲每秒降低計算量
    return wav
```

```python
wave = load_wav_16000_mono(瑋階梯_file)
nwave = load_wav_16000_mono(頻階梯_file)
```

```python
plt.plot(wave)
plt.plot(nwave)
plt.show()
```
